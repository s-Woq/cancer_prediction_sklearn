```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df =  pd.read_csv("./Cancer_Data.csv")
```

```python
df
```

```python
df.columns
```

```python
df1 =  df.drop(["id","Unnamed: 32"], axis=1)
```

```python
df1.isnull().sum()
```

```python
df["diagnosis"].unique()
```

Diagnosis :
 1. M :  Malignant Cancer
 2. B : Benign Cancer

```python
diagnosis_dummies =  pd.get_dummies(df["diagnosis"], prefix= "Category")
diagnosis_dummies =  diagnosis_dummies.drop("Category_B", axis= 1)
```

```python
df_train =  pd.concat([diagnosis_dummies , df1] , axis=1)
```

```python
# df_train = df_train.drop("diagnosis", axis = 1)
df_train.info()
```

Importing Sklearn Libraries

```python
df_train.rename(columns={"Category_M": "Diagnosis_M"}, inplace = True)
```

```python
df_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
#mean_square_error-> Linear,  accuracy_score -> Classify
```

#Splitting Data

```python
X= df_train.drop("Diagnosis_M", axis =1 )
y= df_train["Diagnosis_M"]
```

```python
X_train, X_test , y_train , y_test =  train_test_split(X,y)
```

```python
classifier =  RandomForestClassifier (n_estimators = 100 , random_state =42)
```

```python
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
accuracy  = accuracy_score(y_test,y_pred)
print(f"Accuracy: {accuracy}")
```
